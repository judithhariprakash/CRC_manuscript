```python
CNV = pd.read_csv('CNV/Bedfiles/CNVPerMB_allsamples.tsv',sep='\t',index_col=0)
```

```python
eccDNA = pd.read_csv('eccDNA/Bed_file_alleccDNA/AlleccDNAperMB.tsv',sep='\t',index_col=0)
```

```python
eccDNA.columns
```

```python
for i in eccDNA.index:
    eccDNA.loc[i,'Chrom'] = i.split(':')[0]
```

```python
eDNA_MB = eccDNA.groupby('Chrom').sum()
```

```python
for i in CNV.index:
    CNV.loc[i,'Chrom'] = i.split(':')[0]
```

```python
cnv_MB =CNV.groupby('Chrom').mean()
```

```python
cnv_MB
```

```python
chrom_count =eccDNA.value_counts('Chrom')
```

```python
chrom_count['chr1']
```

```python
df_eDNA = eDNA_MB.copy()
for i in eDNA_MB.index:
    for j in eDNA_MB.columns:
        df_eDNA.loc[i,j] = eDNA_MB.loc[i,j]/chrom_count[i]
    
```

```python
df_eDNA
```

```python
Rvals = {}
Pvals = {}
for nam in cnv_MB.columns:
    if nam!='CRCtx026':
        print (nam)
        regs= linregress(cnv_MB[nam], df_eDNA[nam])
        Rvals[nam] = np.round(regs[2],2)
        Pvals[nam] = np.round(regs[3],2)
        print (np.round(regs[2],2),regs[3])
```

```python
Rvals = {}
for nam in cnv_MB.columns:
    if nam!='CRCtx026':
        print (nam)
        sns.regplot(x = cnv_MB[nam], y =df_eDNA[nam])
        for i in cnv_MB.index:
                plt.text(x=cnv_MB.loc[i,nam],y=df_eDNA.loc[i,nam],s=i)
        regs= linregress(cnv_MB[nam], df_eDNA[nam])
        Rvals[nam] = np.round(regs[2],4)
        txt = ':'.join(('rvalue',str(np.round(regs[2],4))))
        plt.text(x= cnv_MB[nam].min(),y=df_eDNA[nam].max(), s = txt)
        plt.xlabel ('Average CNV per MB')
        plt.ylabel('Total eccDNA')
        plt.title(nam)
        fn = 'eccDNA/Bed_file_alleccDNA/'+nam+'_eccDNAvsCNVperMB_2024.pdf'
        plt.savefig(fn)
        plt.show()
```

```python
coeff = {}
for j in cnv_MB.index:
    coeff[j] ={}
    k = CNV[CNV.Chrom==j]
    l = eccDNA[eccDNA.Chrom==j]
    for nam in cnv_MB.columns:
        if nam!='CRCtx026':
            regs = scipy.stats.linregress(k[nam], l[nam])
            coeff[j][nam] = np.round(regs[2],4)
```

```python
cc = pd.DataFrame(coeff)
```
