```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Miscellaneous/UPregulatedtop500geneswitheccDNA_overlapMSigDBc2.tsv',sep='\t', index_col=0)
```

```python
for n, i in enumerate(df.index):
    df.loc[i,'Position'] = (n+1)*2
```

```python
df1 =df[:20]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,9))

plt.subplots_adjust(left=0.3, right=0.9, bottom=0.3, top=0.9)
plt.xlim(0,0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

sc = ax.scatter(x=df1['k/K'].values,
                       y=df1['Position'].values,  c= -1*(np.log10(df1['FDR q-value'])),s=df1['# Genes in Overlap (k)'].values*2,
                    )
ax.set_yticks(df1.Position)
ax.set_yticklabels(df1.index)
plt.xlabel('Overlap/Total genes', fontsize = 14)
plt.colorbar(sc, aspect=130, fraction=0.01)
plt.savefig('Miscellaneous/MSigdbC2_20genesets.pdf', dpi = 600, bbox_inches = 'tight')
plt.show()
```
