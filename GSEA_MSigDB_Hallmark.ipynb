```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Miscellaneous/oncogenic_pathway_overlap_PLOT.csv', index_col=0)
```

```python
df
```

```python
for n, i in enumerate(df.index):
    df.loc[i,'Position'] = (n+1)*2
```

```python
df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,9))

plt.subplots_adjust(left=0.3, right=0.9, bottom=0.3, top=0.9)
plt.xlim(0,0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

sc = ax.scatter(x=df['k/K'].values,
                       y=df['Position'].values,  c= -1*(np.log10(df['FDR q-value'])),s=df['# Genes in Overlap (k)'].values*5,
                    )
ax.set_yticks(df.Position)
ax.set_yticklabels(df.index)
plt.xlabel('Overlap/Total genes', fontsize = 14)
plt.colorbar(sc, aspect=130, fraction=0.01)
plt.savefig('Miscellaneous/MSigdboncogenic_C2.pdf', dpi = 600, bbox_inches = 'tight')
plt.show()
```
